```python
# Transpose the DataFrame
properties_T = properties.T

# Reset index to get Boroughs as columns
properties_T = properties_T.reset_index()

# Set the first row as the column headers
properties_T.columns = properties_T.iloc[0]

# Drop the first row
properties_T = properties_T.drop(0)
```

```python
# Melt and clean the DataFrame
clean_properties = pd.melt(properties_T, id_vars=['London_Borough', 'ID'], var_name='Date', value_name='Average_price')

# Drop rows with NaN values in the 'Average_price' column
clean_properties = clean_properties.dropna(subset=['Average_price'])

# Convert 'Average_price' to numeric
clean_properties['Average_price'] = pd.to_numeric(clean_properties['Average_price'])

# Display the first few rows of the cleaned DataFrame
clean_properties.head()
```

```python
dfg = dfg.reset_index()
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
def create_price_ratio(d):
    y1998 = float(d['Average_price'][d['Year'] == 1998].iloc[0])
    y2018 = float(d['Average_price'][d['Year'] == 2018].iloc[0])
    ratio = y2018 / y1998
    return [ratio]

```

```python
#Test
create_price_ratio(dfg[dfg['London_Borough'] == 'Camden'])
```

```python
final = {}
for b in dfg['London_Borough'].unique():
    borough = dfg[dfg['London_Borough'] == b]
    final[b] = create_price_ratio(borough)
```

```python
ratios = pd.DataFrame(final)

ratios_T = ratios.T

ratios_T.head()
```

```python
ratios_T = ratios_T.reset_index()
ratios_T.head()
```

```python
ratios_T.rename(columns={'index':'Borough', 0:'Year'}, inplace=True)
ratios_T.head()
```

```python
top_15 = ratios_T.sort_values(by='2018', ascending = False).head(15)

print(top_15)
```

```python
ax = top_15[['Borough', '2018']].plot(kind='bar')

ax.set_xticklabels(top_15.Borough)

plt.title('Top 15 Boroughs with the Highest Increase in House Prices (1998-2018)')
plt.xlabel('Borough')
plt.ylabel('Price Ratio (2018 vs 1998)')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group by London Borough and find the highest average price for each
highest_prices = clean_properties.groupby('London_Borough')['Average_price'].max().reset_index()

# Plotting
plt.figure(figsize=(12, 8))  # Increase the height of the figure
scatter = plt.scatter(highest_prices['Average_price'], highest_prices['London_Borough'], color='skyblue', marker='o')

# Adding labels and title
plt.xlabel('Average Price')
plt.ylabel('London Borough')
plt.title('Highest Average Price for Each London Borough (Scatter Plot)')

# Adding annotations (borough names) next to the dots
for label, x, y in zip(highest_prices['London_Borough'], highest_prices['Average_price'], highest_prices['London_Borough']):
    plt.annotate(label, (x, y), textcoords="offset points", xytext=(5,0), ha='left', fontsize=8)

# Display the plot
plt.grid(axis='x')
plt.show()

```

```python
# Find the row with the highest average price
highest_price_row = clean_properties.loc[clean_properties['Average_price'].idxmax()]

# Display the London Borough and its highest average price
print(f"The London Borough with the highest average price is {highest_price_row['London_Borough']} with a price of £{highest_price_row['Average_price']:.2f}")
```
